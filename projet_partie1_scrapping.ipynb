```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
import time
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
import re
import os
```

```python
browser = webdriver.Chrome('/snap/bin/chromium.chromedriver')

browser.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')

time.sleep(1)
```

```python
browser.maximize_window()
time.sleep(1)
```

```python
liste_film = {
    'title' : [],
    'rating' : [],
    'ranking' : [],
    'director' : [],
    'actors' : [],
    'country' : [],
    'language' : [],
    'date_sortie' : [],
    'production' : [],
    'synopsis' : []
}
```

```python
df_film = pd.DataFrame.from_dict(liste_film)
```

# Algo:
### ouvrir la page.
### pointer sur le premier elt
### cliquer sur l'elt
### récuperer qlq données
### retour
### cliquer sur l'elt en dessous 
### repeat

```python
for i in range(78 , 101):
    action = browser.find_element_by_xpath('//*[@id="main"]/div/span/div/div/div[3]/table/tbody/tr['+str(i+1)+']/td[2]/a')
    action.click()
    time.sleep(1)
    
    titre = browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[1]/div[1]/h1').text
    rating = browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[1]/div[2]/div/div[1]/a/div/div/div[2]/div[1]/span[1]').text
    ranking = i+1
    director = browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/div[3]/ul/li[1]/div/ul/li/a').text
    
    actors = []
    for j in range(1 , 4):
        
        actors.append(browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/div[3]/ul/li[3]/div/ul/li['+str(j)+']').text)
    
    try:
        synopsis = browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/div[1]/p/span[3]').text
    except(NoSuchElementException):
        synopsis = np.NaN
        
    ul_details = browser.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section/div[2]/ul/li/span[contains(text(),"Pays d’origine")]').find_element_by_xpath("./../..").find_elements_by_tag_name('li')
    
    try:
        date_sortie = ul_details[1].text
    except(IndexError):
        date_sortie = np.NaN
    try:        
        country = ul_details[3].text
    except(IndexError):
        country = np.NaN
    try:    
        language = ul_details[8].text
    except(IndexError):
        language = np.NaN
        
    try:
        production = ul_details[14].text
    except(IndexError):
        production = np.NaN
    
    row = {
        'title' : titre,
        'rating' : rating,
        'ranking' : ranking,
        'director' : director,
        'actors' : actors,
        'country' : country,
        'language' : language,
        'date_sortie' : date_sortie,
        'production' : production,
        'synopsis' : synopsis
    }
    df_film = df_film.append(row, ignore_index=True)

    
    browser.back()
```
